```python
# Binary Search

def binarysearch(array,target):
    min = 0
    max = len(array)-1
    
    while min <= max:
        index = (min+max)//2
        
        if array[index]==target:
            return index + 1
        elif array[index]<target:
            min = index+1
        else:
            max = index-1
            
    return -1   

with open("/content/rosalind_bins.txt",'r') as f:
  num = f.read().splitlines()
Array = [int(i) for i in num[2].split()]
targets = [int(i) for i in num[3].split()]
line = ''

for i in range(len(targets)):
  line += str(binarysearch(Array,targets[i])) + ' '

print(line)
```

```python
# Find the Minimum Number of Coins Needed to Make Change

def find_change(coins, money):
    coincount = []
    for coin in coins:
      if money % coin ==0:
        coincount.append(money//coin)

    coin_num = {}
    while money !=0:
      coin_num[max(coins)]=money//max(coins)
      money=money%max(coins)
      coins.remove(max(coins))
    coincount.append(sum(coin_num.values()))
    return min(coincount)


money = 18639
coins = [1,3,5,16,20]
print(find_change(coins,money))
```

```python
# Generate the d-Neiborhood of a string

def intermediateNeighbor(Pattern): #all k-mers with hammington disatance do not exceed 
  mer_set = ['Pattern']
  words = ['A','T','G','C']
  for i in range(len(Pattern)):
    char = Pattern[i]

    for symbol in words:
      if symbol !=char:
        #should replace and add to mers
        new_mer = Pattern[0:1]+symbol+Pattern[i+1]
        mer_set.append(new_mer)
    return mer_set
```

```python
def iterativeNeighbor(Pattern,d):
  mer_set = set([Pattern])
  for i in range(d):
    mer_set_copy = mer_set.copy()
    for mer in mer_set_copy:
      #generate its neighbors
      all_list = set(intermediateNeighbor(mer))
      for a in all_list:
        mer_set.add(all_list)

  return mer_set

iterativeNeighbor('ACG',1)
```
